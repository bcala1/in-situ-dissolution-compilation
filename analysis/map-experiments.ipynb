## Figure 3

```python
plt.figure(dpi=300)
usemap_proj = ccrs.PlateCarree(central_longitude=180)
ax = plt.axes(projection=usemap_proj)
ax.stock_img()
ax.set_extent([130, 359, -20, 70], crs=ccrs.PlateCarree())

plt.scatter(lons, lats, color='red', s=20, marker="x", linewidths=1, transform=ccrs.PlateCarree())

plt.text(p66_lon-7, p66_lat -10, 'Peterson (1966), \nBerger (1967)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(m77_lon_1-1.5, m77_lat_1 -7, 'Milliman (1977)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(he78_lon +3, he78_lat -1, 'Honjo & Erez (1978)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(t81_lon-6, t81_lat -10, 'Thunell et al. \n(1981)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(m82_lon1-6, m82_lat1 -10, 'Metzler et al. \n(1982)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(t97_lon+3, t97_lat -1, 'Troy et al. (1997)', fontsize=6,
         transform=ccrs.PlateCarree())

plt.text(f08_lon-8, f08_lat -10, 'Fukuhara et al.\n(2008)', fontsize=6,
         transform=ccrs.PlateCarree())


plt.text(d19_lon4+10, d19_lat4 -7, 'Dong et al.(2019), \nNaviaux et al.(2019), \nSubhas et al.(2022)', fontsize=6,
         transform=ccrs.PlateCarree())

arrow1 = mpatches.FancyArrowPatch((-147, 39), (-140, 39),
                                   arrowstyle = mpatches.ArrowStyle("]-", widthA=22.0, lengthA=1,angleA=None), 
                                  lw=0.8, transform=ccrs.PlateCarree())

ax.add_patch(arrow1)




gl = ax.gridlines(draw_labels=True, crs=ccrs.PlateCarree(), color='gray', linewidth=0.3)


gl.xlabel_style = {'size': 6}
gl.ylabel_style = {'size': 6}
```
